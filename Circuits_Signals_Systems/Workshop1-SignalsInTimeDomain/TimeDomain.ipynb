# 1.1 Real CT Sinusoid

General Sinusoidal signal
$$x(t) = A cos(\omega_o t +\phi) $$

Where:
$$\omega_o = 2\pi f_o$$
$$\phi = phase





```python
# Importing Required libraries
import numpy as np
import matplotlib.pyplot as plt

# Sampling Frequency
fs = 44100
ts = 1/fs
# A linearly-spaced array with step ts
t = np.arange(-1., 1., ts)

```

$$x(t) = A cos(\omega_o t +\phi) $$
When:
$$\phi = 0$$

```python
A = 1
f = 2
w0 =2 * np.pi * f

phi = 0
xt = A *np.cos(w0 * t + phi)
fig, ax = plt.subplots()

ax.grid(True)
ax.plot(t, xt)

plt.show()

```

### Task 1
Generatng a real sinusoidal signals:
1. $$x_1(t) =  cos(2  \pi t) $$
2. $$x_2(t) = 0.5 cos(2  \pi t) $$
3. $$x_3(t) =  cos(2  \pi t + \pi/2) = -sin(2  \pi t) $$
4. $$x_4(t) =  cos(4  \pi t )$$


```python
fig, axes = plt.subplots( 2, 2, sharex = "all", sharey='all', figsize=(18, 18))

A = 1
f = 1
w0 =2 * np.pi * f
phi = 0
xt1 = A *np.cos(w0 * t + phi)

A = 0.5
f = 1
w0 =2 * np.pi * f
phi = 0
xt2 = A *np.cos(w0 * t + phi)

A = 1
f = 1
w0 =2 * np.pi * f
phi = np.pi/2
xt3 = A *np.cos(w0 * t + phi)

A = 1
f = 2
w0 =2 * np.pi * f
phi = 0
xt4 = A *np.cos(w0 * t + phi)

axes[0, 0].plot(t, xt1)
axes[0, 0].set_xlabel(r'$t$')
axes[0, 0].set_ylabel(r'$xt1$')
axes[0, 0].grid(True)
axes[0, 0].title.set_text(r'$cos(2 \pi t)$')

axes[0, 1].plot(t, xt2)
axes[0, 1].set_xlabel(r'$t$')
axes[0, 1].set_ylabel(r'$xt2$')
axes[0, 1].grid(True)
axes[0, 1].title.set_text(r'$0.5\cos(2\pi t)$')

axes[1, 0].plot(t, xt3)
axes[1, 0].set_xlabel(r'$t$')
axes[1, 0].set_ylabel(r'$xt3$')
axes[1, 0].grid(True)
axes[1, 0].title.set_text(r'$-\sin(2\pi t)$')

axes[1, 1].plot(t, xt4)
axes[1, 1].set_xlabel(r'$t$')
axes[1, 1].set_ylabel(r'$xt4$')
axes[1, 1].grid(True)
axes[1, 1].title.set_text(r'$\cos(4\pi t)$')


plt.show()
```

Here, `x1`, `x2` and `x4` are Even signals and `x3` is an odd signal.

# 1.1 Real CT Exponential
$$x(t) = C \exp^{\alpha*t}$$


```python
fs = 4.4e4
ts = 1/fs
t = np.arange(-1., 1., ts)

C = 1.0
alpha = 1.2

xt = C * np.exp(alpha * t)

fig, ax = plt.subplots()
ax.plot(t, xt, label=r'$C\exp(\alpha t$')
ax.set_xlabel(r'$t$')
ax.set_ylabel(r'$x(t)$')
ax.grid(True)
ax.legend(loc='lower right')

plt.show()


```

When:
$$  \alpha < 0$$
x(t) becomes a decaying exponential


```python
fs = 4.4e4
ts = 1/fs
t = np.arange(-1., 1., ts)

C = 1.0
alpha = -1.2

xt = C * np.exp(alpha * t)

fig, ax = plt.subplots()
ax.plot(t, xt, label=r'$C\exp(\alpha t$')
ax.set_xlabel(r'$t$')
ax.set_ylabel(r'$x(t)$')
ax.grid(True)
ax.legend(loc='lower right')

plt.show()
```

# Growing sinusoidal Signal
$$x(t) = A \exp^{\alpha*t} cos(\omega_o t +\phi) $$





1. When C = 0.15, r = 2, $\theta = 0$

```python
fs = 4.4e4
ts = 1/fs
t = np.arange(-1., 1., ts)
f = 5

w0 =2 * np.pi * f
C = 0.15
r = 2
theta = 0

xt = C * np.exp(r * t) * np.cos(w0 * t + theta)

fig, ax = plt.subplots()
ax.plot(t, xt, label=r'$\exp^{\alpha*t} cos(\omega_o t)$')
ax.set_xlabel(r'$t$')
ax.set_ylabel(r'$x(t)$')
ax.grid(True)
ax.legend(loc='lower left')

plt.show()
```

```python
fs = 4.4e4
ts = 1/fs
t = np.arange(-1., 1., ts)
f = 5

w0 =2 * np.pi * f
C = 0.15
r = -2
theta = np.pi/2

xt = C * np.exp(r * t) * np.cos(w0 * t + theta)

fig, ax = plt.subplots()
ax.plot(t, xt, label=r'$\exp^{\alpha*t} cos(\omega_o t + \theta)$')
ax.set_xlabel(r'$t$')
ax.set_ylabel(r'$x(t)$')
ax.grid(True)
ax.legend(loc='lower left')

plt.show()
```

# 1.6 General Complex Exponential Signals
Plotting the real part of 
$$x[n] = C \alpha^n$$
with 
$$ C= |C| e^{j\theta}$$
$$ \alpha = |\alpha| e^{j\omega_0}$$

Hint: 
Re{x[n]} = |C| |$\alpha$|<sup>n</sup> cos($\omega_0$ n + $\theta$)

When:
1. C = 1, $\alpha = 1.1, \omega_o = \frac{2\pi}{12}$
2. C = 1, $\alpha = 0.92, \omega_o = \frac{2\pi}{12}$


```python
n = np.arange(-12, 12, 1)
fig, axes = plt.subplots( 2, 1, sharex='all', sharey='all', figsize=(18, 18))
theta = 0

c = 1
alpha = 1.1
w0 = 2 * np.pi /12
Rxn1 = np.absolute(c) * np.absolute(alpha)**n * np.cos(w0 * n + theta)

c = 1
alpha = 0.92
w0 = 2 * np.pi /12
Rxn2 = np.absolute(c) * np.absolute(alpha)**n * np.cos(w0 * n + theta)

axes[0].stem(n, Rxn1)
axes[0].set_xlabel(r'$n$')
axes[0].set_ylabel(r'$Rxn1$')
axes[0].grid(True)
axes[0].title.set_text(r'$\alpha = 1.1$')

axes[1].stem(n, Rxn2)
axes[1].set_xlabel(r'$n$')
axes[1].set_ylabel(r'$Rxn2$')
axes[1].grid(True)
axes[1].title.set_text(r'$\alpha = 0.92$')

plt.show()
```

What happens when $\alpha$ is a negative number?


```python
n = np.arange(-12, 12, 1)
fig, axes = plt.subplots( 2, 1, sharex='all', sharey='all', figsize=(18, 18))
theta = 0

c = 1
alpha = -1.1
w0 = 2 * np.pi /12
Rxn1 = np.absolute(c) * np.absolute(alpha)**n * np.cos(w0 * n + theta)

c = 1
alpha = -0.92
w0 = 2 * np.pi /12
Rxn2 = np.absolute(c) * np.absolute(alpha)**n * np.cos(w0 * n + theta)

axes[0].stem(n, Rxn1)
axes[0].set_xlabel(r'$n$')
axes[0].set_ylabel(r'$Rxn1$')
axes[0].grid(True)
axes[0].title.set_text(r'$\alpha = 1.1$')

axes[1].stem(n, Rxn2)
axes[1].set_xlabel(r'$n$')
axes[1].set_ylabel(r'$Rxn2$')
axes[1].grid(True)
axes[1].title.set_text(r'$\alpha = 0.92$')

plt.show()
```

Real part of the signal doesn't change, when $\alpha$ is negative. The signal is still a exponential.

# 1.7 Transformation of the Independent Variable
$$ x(t)=   \left\{
\begin{array}{ll}
      0 & t<0 \\
      1 & 0<t<1 \\
      2-t & 1<t<2 \\
      0 & t > 2
\end{array} 
\right.  $$


```python
def x(t):
    if t<0.:
        return 0.
    elif t<1.:
        return 1.
    elif t<2.:
        return 2.-t
    else:
        return 0.

fs = 4.4e4
ts = 1/fs
t = np.arange(-3.5, 3.5, ts)
fig, axes = plt.subplots( 7, 1, sharey='all', figsize=(10, 15))
fig.tight_layout(pad=3.0)

xt = np.array([x(t) for t in t])
axes[0].plot(t, xt)
axes[0].set_xlabel(r'$t$')
axes[0].set_ylabel(r'$x(t)$')
axes[0].grid(True)
axes[0].title.set_text(r'$x(t)$')

xt = np.array([x(t-1) for t in t])
axes[1].plot(t, xt)
axes[1].set_xlabel(r'$t$')
axes[1].set_ylabel(r'$x(t-1)$')
axes[1].grid(True)
axes[1].title.set_text(r'$x(t-1)$')

xt = np.array([x(t+1) for t in t])
axes[2].plot(t, xt)
axes[2].set_xlabel(r'$t$')
axes[2].set_ylabel(r'$x(t+1)$')
axes[2].grid(True)
axes[2].title.set_text(r'$x(t+1)$')

xt = np.array([x(-t) for t in t])
axes[3].plot(t, xt)
axes[3].set_xlabel(r'$t$')
axes[3].set_ylabel(r'$x(-t)$')
axes[3].grid(True)
axes[3].title.set_text(r'$x(-t)$')

xt = np.array([x(-t+1) for t in t])
axes[4].plot(t, xt)
axes[4].set_xlabel(r'$t$')
axes[4].set_ylabel(r'$x(-t+1)$')
axes[4].grid(True)
axes[4].title.set_text(r'$x(-t+1)$')

xt = np.array([x(3*t / 2) for t in t])
axes[5].plot(t, xt)
axes[5].set_xlabel(r'$t$')
axes[5].set_ylabel(r'$x(3t/2)$')
axes[5].grid(True)
axes[5].title.set_text(r'$x(3t/2)$')

xt = np.array([x(3*t/2 + 1) for t in t])
axes[6].plot(t, xt)
axes[6].set_xlabel(r'$t$')
axes[6].set_ylabel(r'$x(3t/2+1)$')
axes[6].grid(True)
axes[6].title.set_text(r'$x(3t/2+1)$')


plt.show()
```

# Observing a signal in Frequency Domain
Plotting a signal and its frequency domain representation
Note: Non-zero values are actually delta functions in frequency domain.


```python
fs = 100
ts = 1/fs
t = np.arange(-1., 1., ts)
fig, axes = plt.subplots(2, 1, figsize=(10, 10))

f = 2
w0 =2 * np.pi * f

xt = 0.75 + 2*np.cos(w0 *t) + 1.*np.cos(3*w0*t)

Xf = np.fft.fft(xt)
freq = np.fft.fftfreq(t.shape[-1], d=ts)

axes[0].plot(t, xt)
axes[0].set_xlabel(r'$t$')
axes[0].set_ylabel(r'$x(t)$')
axes[0].grid(True)
axes[0].title.set_text(r'$x(t)$')

valsubrange = np.concatenate((np.arange(0, 21, 1), np.arange(-1, -21, -1)))
freqsubrange = np.concatenate((np.arange(0, 21, 1), np.arange(-1, -21, -1)))
axes[1].stem(freq[freqsubrange], Xf.real[valsubrange]/len(t))

axes[1].set_xlabel(r'$f$ in Hz')
axes[1].set_ylabel(r'$\mathfrak{Re}(X(j\omega))$')

plt.xticks(np.arange(-10,11))
plt.show()
```
