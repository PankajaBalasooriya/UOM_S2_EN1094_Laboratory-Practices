# 1.1 Real CT Sinusoid

General Sinusoidal signal
$$x(t) = A cos(\omega_o t +\phi) $$

Where:
$$\omega_o = 2\pi f_o$$
$$\phi = phase





```python
# Importing Required libraries
import numpy as np
import matplotlib.pyplot as plt

# Sampling Frequency
fs = 44100
ts = 1/fs
# A linearly-spaced array with step ts
t = np.arange(-1., 1., ts)

```

$$x(t) = A cos(\omega_o t +\phi) $$
When:
$$\phi = 0$$

```python
A = 1
f = 2
w0 =2 * np.pi * f

phi = 0
xt = A *np.cos(w0 * t + phi)
fig, ax = plt.subplots()

plt.show()

```

```python
fig, axes = plt.subplots( 2, 2, sharex = "all", sharey='all', figsize=(18, 18))

A = 1
f = 1
w0 =2 * np.pi * f
phi = 0
xt1 = A *np.cos(w0 * t + phi)

A = 0.5
f = 1
w0 =2 * np.pi * f
phi = 0
xt2 = A *np.cos(w0 * t + phi)

A = 1
f = 1
w0 =2 * np.pi * f
phi = np.pi/2
xt3 = A *np.cos(w0 * t + phi)

A = 1
f = 2
w0 =2 * np.pi * f
phi = 0
xt4 = A *np.cos(w0 * t + phi)

axes[0, 0].plot(t, xt1)
axes[0, 0].set_xlabel('$t$')
axes[0, 0].set_ylabel('$xt1$')
axes[0, 0].grid(True)
axes[0, 0].title.set_text('$cos(2 \pi t)$')

axes[0, 1].plot(t, xt2)
axes[0, 1].set_xlabel('$t$')
axes[0, 1].set_ylabel('$xt2$')
axes[0, 1].grid(True)
axes[0, 1].title.set_text('$\cos(2\pi t)$')

axes[1, 0].plot(t, xt3)
axes[1, 0].set_xlabel('$t$')
axes[1, 0].set_ylabel('$xt3$')
axes[1, 0].grid(True)
axes[1, 0].title.set_text('$\cos(2\pi t)$')

axes[1, 1].plot(t, xt4)
axes[1, 1].set_xlabel('$t$')
axes[1, 1].set_ylabel('$xt4$')
axes[1, 1].grid(True)
axes[1, 1].title.set_text('$\cos(4\pi t)$')

#axes.plot(t, xt)

plt.show()
```
